# Data Cleaning

```python
df_clean=data.copy()
```

```python
df_clean['medical_specialty']=df_clean['medical_specialty'].replace('Missing','Unknown')
```

```python
df_clean['diag_1']=df_clean['diag_1'].replace('Missing',np.nan)
df_clean['diag_2']=df_clean['diag_2'].replace('Missing',np.nan)
df_clean['diag_3']=df_clean['diag_3'].replace('Missing',np.nan)
```

```python
df_clean['diag_3'].unique()
```

```python
df_clean.dropna(subset = ['diag_1','diag_2','diag_3'],inplace=True)
```

```python
df_clean['diag_1'].isnull().sum()
```

```python
import os
import pandas as pd

folder_path = os.path.join("..", "data")
file_name = "hospital_readmission_clean.csv"
file_path = os.path.join(folder_path, file_name)

os.makedirs(folder_path, exist_ok=True)

if not os.path.exists(file_path):
    df_clean.to_csv(file_path, index=False)
```
